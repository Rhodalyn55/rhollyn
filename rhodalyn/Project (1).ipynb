```python
#Task 1
#Create a dictionary from the given data set
years = [i for i in range(2011,2021)] 
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
movie_dict = {'years': years, 'durations' : durations}
movie_dict 
```

```python
#Task 2
#Create a dataframe from the dictionary created
import pandas as pd 
durations_df = pd.DataFrame(movie_dict)
durations_df 
```

```python
#Task 3
#Visual inspection of our data
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(durations_df['years'], durations_df['durations'])
plt.title('Netflix Movie Durations 2011-2020')
plt.plot()
```

```python
#Task 4
#Creating another dataframe from the CSV file
import pandas as pd
netflix_df = pd.read_csv('netflix_data.csv')
netflix_df.head(5)
```

```python
#Task 5
#Creating subsets from our created dataframe
netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]
netflix_movies_col_subset = netflix_df[["title", "country", "genre", "release_year", "duration"]]
netflix_movies_col_subset.head(5)
```

```python
#Task 6
#Creation of scatter plot of netflix_movies_col_subset
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])
plt.xlabel("release_year")
plt.ylabel("duration(min)")
plt.title("Movie duration by year of release")
fig = plt.figure(figsize=(12,8))
plt.show()
```

```python
#Task 7
#Filter dataframe to show movie duration less than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]
short_movies.head(20)
```

```python
#Task 8
#Marking Non-feature films
colors = []
for idx, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-up': 
        colors.append('green')
    else: 
        colors.append('black')
netflix_movies_col_subset.head(10) 
```

```python
#Task 9
#plot scatter plot of data contained in netflix_movies_col_subset with color
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], color=colors)
plt.title("Movie duration by year of release")
plt.xlabel("Release Year")
plt.ylabel("Duration(min)")
plt.show()
```
